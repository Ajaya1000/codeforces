```python
def give(n: int):
    curr = n
    first = True
    count = 0
    while first or curr > n:
        first = False
        if curr%2 == 1:
            curr = 3*curr + 1
        while curr%2 == 0:
            curr =  curr // 2
        count += 1
    return (curr,count)
```

```python
n = 100
for i in range(1,n):
    print(f"for {i}: {give(i)}")
```
